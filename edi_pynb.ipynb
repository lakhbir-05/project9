```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/content/winequality-red12.txt', delimiter=";") # delimiter(sperartor) between values is semicolon
print("first five rows:")
df.head(5)
```

```python
print("last five rows:")
df.tail(5)
```

```python
print("column info:")
df.info()
```

```python
print("shape of dataset:", df.shape)
print(df.columns)
print(df.keys())
```

```python
print("summary stats:")
print(df.describe())
```

```python
pd.unique(df['quality'])
```

```python
df= df.drop(["alcohol","chlorides"], axis=1) # dropping irrelevent columns
df.head()
```

```python
# renaming columns
df= df.rename(columns={'fixed acidity':'fixed_acidity', 'volatile acidity':'volatile_acidity', 'citric acid':'citric_acid', 'residual sugar':'residual_sugar', 'free sulfur dioxide':'free_sulfur_dioxide', 'total sulfur dioxide':'total_sulfur_dioxide'})
df.head()
```

```python
# dropping duplicate rows
print(df.shape)
duplicate_rows= df[df.duplicated()]
print("number of duplicate rows:", len(duplicate_rows))
print(df.head(2))
print("after deleting duplicate rows the shape of df is:",df.shape)
```

```python
df = df.drop_duplicates()
df.head()
```

```python
df.count()
```

```python
# dropping the missing or null values
print(df.isnull().sum())
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
print(df.isnull().sum())   # After dropping the values
```

```python
sns.boxplot(x=df['quality'])
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['pH'])
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['density'], y=df['quality'])
plt.title("scatter plot between density and quality")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=df['citric_acid'],y=df['sulphates'])
plt.title(" line plot between citric_acid and sulphates")
plt.show()
```

```python
line plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("feature correlation heat map")
plt.show()
```
